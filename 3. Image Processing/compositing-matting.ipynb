```python
# b_image = cv2.imread('../flowers/daisies_00028.jpg')
f_image = cv2.imread('../flowers/daisies_00039.jpg')

b_image = cv2.imread('../rainbow.jpg', 1)

print(f_image.shape, b_image.shape)

plt.figure(figsize=(10,6))

plt.subplot(2,1,1)
plt.imshow(f_image)

plt.subplot(2, 1,2)
plt.imshow(b_image)
```

```python
# reshaping both images to the same pixels 
f_image = my_cv.resize_and_pad(f_image, size=(800, 800), pad_color=255)
b_image = my_cv.resize_and_pad(b_image, size=(800, 800), pad_color=255)

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.imshow(f_image)

plt.subplot(1,2,2)
plt.imshow(b_image)
```

```python
f_mask = my_cv.binary_mask(f_image)
b_mask = my_cv.binary_mask(b_image)

plt.figure(figsize=(10,6))

plt.subplot(1,2,1)
plt.imshow(f_mask)

plt.subplot(1,2,2)
plt.imshow(b_mask)
```

```python
f_r, f_g, f_b = cv2.split(f_image)
b_r, b_g, b_b = cv2.split(b_image)
```

```python
# for opacity of the foreground image
alpha = 0.2
new_image = (1-alpha)*b_image + (alpha*f_image)
plt.imshow(new_image.astype(int))
```
