```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("../flower.png")
plt.imshow(img)
```

```python
annotated_img = img.copy() 
cv2.rectangle(annotated_img, pt1=(580,440), pt2=(980, 780), color=(255,0,0), thickness=6)
plt.imshow(annotated_img)
```

```python

f, axes = plt.subplots(1,3, figsize=(15,5))

mask = np.zeros(img.shape[:2], np.uint8)
axes[0].imshow(mask)

mask[440:780, 580:980] = 255
masked_img = cv2.bitwise_and(img, img, mask=mask)


axes[1].imshow(mask)
axes[2].imshow(masked_img)
plt.show() 
# plt.imshow(mask, cmap='gray')
```

```python
img_red_hist = cv2.calcHist([img], [0], None, [256], [0,256])
img_green_hist = cv2.calcHist([img], [1], None, [256], [0,256])
img_blue_hist = cv2.calcHist([img], [2], None, [256], [0,256])

f, axes = plt.subplots(1,1, figsize=(12,6))

axes.plot(img_red_hist, color='red')
axes.plot(img_green_hist, color='green')
axes.plot(img_blue_hist, color='blue')
plt.show() 
```

```python
mask_hist = cv2.calcHist([img], [0], None, [256], [0,256])
masked_img_hist = cv2.calcHist([img], [0], mask, [256], [0,256])

f, axes = plt.subplots(1,3, figsize=(15,5))

axes[0].plot(mask_hist)
axes[1].plot(masked_img_hist, color='orange')
axes[2].hist(img.flatten(), 256, [0,256], 'orange')
plt.show() 
# plt.imshow(mask, cmap='gray')
```
