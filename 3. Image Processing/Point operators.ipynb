```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread(cv2.samples.findFile("../flowers/daisies_00025.jpg"))
plt.imshow(img)
```

```python
img.shape 
```

```python
def create_histogram(color_channel):
    hist = np.zeros(256)
    for pixel in  color_channel.flatten():
        hist[pixel] += 1 
    return hist
```

```python
# splitting an image
r,g,b = cv2.split(img)
r.shape,  g.shape, b.shape 
```

```python
hist_r = create_histogram(r)
hist_g = create_histogram(g)
hist_b = create_histogram(b)
hist_r.shape, hist_b.shape, hist_g.shape 
```

```python
hist_r = cv2.calcHist([r], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([g], [0], None, [256], [0, 256])
hist_b = cv2.calcHist([b], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(10,6))

plt.imshow(img)
plt.title('flower')
plt.axis('off')

plt.subplot(2,1,2)
plt.plot(hist_r, color='r', label='Red')
plt.plot(hist_g, color='g', label='Green')
plt.plot(hist_b, color='b', label='Blue')

plt.legend()
plt.show() 
```
